```python
####################################################
import cv2
import numpy as np

kernel = np.ones((1, 1), np.uint8)
def read_file(filename):
    img = cv2.imread(filename)
    cv2_imshow(img)
    return img

def color_quantization(img, k):
# Transform the image
    data = np.float32(img).reshape((-1, 3))
    # Determine criteria
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 0.001)

    # Implementing K-Means
    ret, label, center = cv2.kmeans(data, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
    center = np.uint8(center)
    result = center[label.flatten()]
    result = result.reshape(img.shape)
    return result

def edge_mask(img, line_size, blur_value):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_blur = cv2.medianBlur(gray, blur_value)
    edges = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, line_size, blur_value)
    return edges

line_size = 3
blur_value =3

img = cv2.imread("dhoom.jpg")
# img = cv2.medianBlur(img, 3)

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply adaptive thresholding to convert the image to binary
gray = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 19, 19)



edges = edge_mask(img, line_size, blur_value)
total_color = 100
img = color_quantization(img, total_color)
blurred = cv2.bilateralFilter(img, d=1, sigmaColor=350,sigmaSpace=350)
# cv2_imshow(blurred)
# cv2.imshow("img",blurred)

cartoon = cv2.bitwise_and(blurred, blurred, mask=edges)

color = cv2.bilateralFilter(img, 5 ,350,350)

cartoon = cv2.bitwise_and(color, color, mask=gray)
img_erosion = cv2.erode(cartoon, kernel, iterations=2)
# cv2.imshow("img",cartoon)
cv2.imwrite('dhoom_new1.jpg',img_erosion)
# cv2.imwrite('sud_sid_new.jpg',cartoon)
print("Done")
cv2.waitKey(0)
cv2.destroyAllWindows()
```
